```python
def generate_dict_lemmas(path = "generate/generate.txt", encode = 'unicode_escape'):
  keys = [] 
  values= []
  arr = []
  with open(path, 'r',encoding = encode) as file:
    lines = file.read()
  lines = lines.split("\n")
  file.close()
  for line in lines:
    arr = line.strip().split(" ")
    if len(arr) >2:
      keys.append(arr[0].replace(".",'').replace("#",''))
      values.append(arr[-1])
  dictionary = dict(zip(keys, values))
  return dictionary

def tokenization_sents(text):
  sentences = nltk.sent_tokenize(text)
  return sentences


def tokenization_word_tokenize(text, is_lower_case=True):
  if is_lower_case:
    tokenize = word_tokenize(text.lower())
  else:
    tokenize = word_tokenize(text)
  return tokenize

def remove_special_characters(words):
  return [w.lower() for w in words if w.isalpha()]


def lemmatization(dict_lemmas, word):
  if word in dict_lemmas:
    return dict_lemmas[word]
  else:
    return word

def remove_stopwords(words, stopwords = spanish_stopwords):
  if len(words) > 0:
    if type(words[0]) is tuple:
      content = [(w,t) for (w,t) in words if w.lower() not in stopwords]
    else:
      content = [w for w in words if w.lower() not in stopwords]
    return content
  else:
    return []

def normalization(text, dict_lemmas):
  normalize_sents = []
  sents= tokenization_sents(text)
  for sent in sents:
    words = tokenization_word_tokenize(sent)
    words = remove_special_characters(words)
    words = [lemmatization(dict_lemmas, w) for w in words]
    words = remove_stopwords(words)
    normalize_sents.append(words)
  return normalize_sents


def print_results(file_name, sorted_dict):
  with open(file_name+'.txt', 'w') as f:
    for (w, v) in sorted_dict:
      f.write( w +' : ' + str(v) + '\n')
  f.close()

def flatten_text(texts):
  texts_ = []
  for t in texts:
    tt = ''
    for sent in t:
      text = ' '.join(sent)
      if tt == '':
        tt = text
      else: 
        tt = tt + ' ' + text 
    texts_.append(tt)
  return texts_

def flatten_corpus(corpus): 
  return ' '.join([document.strip() for document in corpus])


def compute_ngrams(sequence, n):
    return zip(*[sequence[index:] for index in range(n)])

def get_top_ngrams(corpus, ngram_val=1, limit=5):
    corpus = flatten_corpus(corpus)
    tokens = nltk.word_tokenize(corpus)
    ngrams = compute_ngrams(tokens, ngram_val)
    ngrams_freq_dist = nltk.FreqDist(ngrams)
    sorted_ngrams_fd = sorted(ngrams_freq_dist.items(),key=itemgetter(1), reverse=True)
    sorted_ngrams = sorted_ngrams_fd[0:limit]
    sorted_ngrams = [(' '.join(text), freq)
                     for text, freq in sorted_ngrams]
    return sorted_ngrams
```
